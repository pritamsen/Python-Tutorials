Let's import the Pandas and Seaborn libraries, and the `pyplot` module of the Matplotlib library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's load and explore the dataset

```python
df = pd.read_csv('heart.csv')
df
```

```python
df.info()
```

Let's make a basic scatter plot

```python
# Using matplotlib
plt.scatter(data = df, x = 'age', y = 'cholestrol')
```

The object and it's location in memory is being displayed. We can avoid this in two ways.

```python
plt.scatter(data = df, x = 'age', y = 'cholestrol');
```

```python
plt.scatter(data = df, x = 'age', y = 'cholestrol')
plt.show()
```

Let's change the size of the image, and also annotate the plot by adding a title and giving names to the axes

```python
plt.figure(figsize = (5, 3))
plt.scatter(data = df, x = 'age', y = 'cholestrol')
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

```python
# Using seaborn
sns.scatterplot(data = df, x = 'age', y = 'cholestrol')
```

```python
# Seaborn is built atop matplotlib; many matplotlib methods are compatible with it
plt.figure(figsize = (5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol')
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

You can use parameters to customise your plot and make it more interpretable and/or aesthetically pleasing

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', s = 10)  # change the size of the markers
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', marker = '^')  # change the style of the markers
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', c = 'r')  # change the colour of the markers
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', alpha = 0.5)  # change the opacity of the markers
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

Let's apply all of these changes together

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', s = 10, marker = '^', c = 'r', alpha = 0.5)
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

You may want to change the style of the plots. For instance, the following changes the plot to look more like the plots one obtains by using the ggplot library of the R programming language. This style is often used in academic publications.

```python
plt.style.use('ggplot')  # changing the plot style to 'ggplot'
plt.figure(figsize = (5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', s = 10, marker = '^', c = 'r', alpha = 0.5)
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```

Style changes are universal and will affect all plots. You need to manually revert the style if they were intended for just one plot.

```python
plt.style.use('default')  # reverting to the default style
plt.figure(figsize = (5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', s = 10, marker = '^', c = 'r', alpha = 0.5)
plt.title('Plot of age versus cholesterol levels')
plt.xlabel('Cholesterol')
plt.ylabel('Age')
plt.show()
```
